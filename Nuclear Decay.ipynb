*Nuclear Decay* by Euler Method

```python
import math as m
import matplotlib.pyplot as plt
%matplotlib inline
ti=0.0
tf=10.0
tao=1
N0=10000
dt=0.1
t=[ti]
N=[N0]
n=int((tf-ti)/dt)
for j in range (n):
    N.append(N[j]-N[j]/tao * dt)
    t.append(t[j]+dt)
    
#1: Plot Nuclear decay by Euler Method
plt.figure(1)
plt.xlabel("t")
plt.ylabel("N")
plt.plot(t,N,'k',label='Euler')
plt.legend()

#2: Plot for different dt & Compare with Exact Value
NExact=[N0]
plt.figure(2)
plt.xlabel("t")
plt.ylabel("N")
dt=0.5
t=[ti]
N=[N0]
n=int((tf-ti)/dt)
for j in range (n):
    N.append(N[j]-N[j]/tao * dt)
    t.append(t[j]+dt)
plt.plot(t,N,'k',label='dt=0.5') 

dt=0.25
t=[ti]
N=[N0]
n=int((tf-ti)/dt)
for j in range (n):
    N.append(N[j]-N[j]/tao * dt)
    t.append(t[j]+dt)
plt.plot(t,N,'r',label='dt=0.25')

dt=0.1
t=[ti]
N=[N0]
n=int((tf-ti)/dt)
for j in range (n):
    N.append(N[j]-N[j]/tao * dt)
    t.append(t[j]+dt)
plt.plot(t,N,'m',label='dt=0.1')

dt=0.05
t=[ti]
N=[N0]
n=int((tf-ti)/dt)
for j in range (n):
    N.append(N[j]-N[j]/tao * dt)
    t.append(t[j]+dt)
plt.plot(t,N,'y',label='dt=0.05')

n=int((tf-ti)/dt)
for j in range (n):
    NExact.append(N0 * m.exp(-t[j]/tao))
plt.plot(t,NExact,'b',label='Exact')
plt.legend()

#3 Error of Euler Method 
plt.figure(3)
plt.xlabel("dt")
plt.ylabel("N-Exact")
dt=0.1
n=int((tf-ti)/dt)
NExact1=[]
N1=[]
dt1=[ti]
NError=[]
tf=[]
for j in range (n):
    dt1.append((j+1)/10000)
    tf.append(ti+(n+1)*dt1[j])
    n=int((tf[j]-ti)/dt1[j+1])
    NExact1.append(N0*m.exp(-tf[j]/tao))
    N1.append(N[n]-N[n]/tao *dt1[j])
    NError.append(abs(NExact1[j]-N1[j]))
del dt1[100]
plt.plot(dt1,NError,'b',label='Error')
plt.legend()
```
